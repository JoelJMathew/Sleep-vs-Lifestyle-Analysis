# Sleep and Lifestyle Choices

# Introduction:

This data analysis dives into the exploration of factors that affect sleep quality and duration. The objective of this analysis is to go over key factors that can affect sleep quality and duration such as career, stress levels, and physical activity levels. To begin this analysis, we will clean the data, and then perform exploratory data analysis to help us see relationships and patterns.

By completing such analysis, I seek to have found enough correlation within the dataset to answer the question: “Do lifestyle choices affect sleep quality and duration?”. 






```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data then display first twenty columns 
data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
data.head(20)
```

```python
# Let's check if cleaning is needed. 
print(data.isnull().sum())
data = data.drop_duplicates()
```

```python
# We can see that there are 219 NaN values for "Sleep disorder" column. 
# We should clean the data and replace "NaN" with "None"
```

```python
data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
# Replace NaN (or missing values) with "None" using the .replace() method
data.replace(to_replace=np.nan, value="None", inplace=True)

# Save changes
data.to_csv('Sleep_health_and_lifestyle_dataset.csv', index=False)
print("NaN values have been replaced with 'None'.")

# Display the first twenty columns of cleaned data
data.head(20)
```

# Key Observations:

Occupation vs Sleep Duration:                                      

```python
# Bar graph for sleep duration vs occupation
avg_sleep_by_occupation = data.groupby('Occupation')['Sleep Duration'].mean().sort_values()
avg_sleep_by_occupation.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Sleep Duration vs Occupation (Fig 1.)', fontsize=12)
plt.xlabel('Occupation', fontsize=12)
plt.ylabel('Average Sleep Duration (hours)', fontsize=12)
plt.show()
```

Does one's occupation affect how many hours they sleep? This graph shows that engineers get the most sleep. Sales representatives get the least amount of sleep. Interestingly, doctors and nurses get around the same amount of sleep per night. We can see that two occupations in the same industry have almost identical sleep durations. This is an important observation as it ties itself to the aspect of occupations affecting sleep duration. 

Occupation vs Stress Level Bar Graph:

```python
# Bar graph for stress level vs occupation 
avg_stress_by_occupation = data.groupby('Occupation')['Stress Level'].mean().sort_values()
avg_stress_by_occupation.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Average Stress Level vs Occupation (Fig 2.)', fontsize=12)
plt.xlabel('Occupation', fontsize=12)
plt.ylabel('Average Stress Level', fontsize=12)
plt.show()
```

This graph shows the relationship between one's occupation and stress level. It seems as if sales representatives are the most stressed. This logically makes sense because sales is one of the most emotionally demanding industries. This is an important observation as stress is a major factor that affects sleep quality. 


Physical Activity Levels vs Stress Levels

```python
# Bar graph for stress level vs physical activity level
avg_stress_by_activity = data.groupby('Physical Activity Level')['Stress Level'].mean()
avg_stress_by_activity.plot(kind='bar', color='coral', edgecolor='black')
plt.title('Average Stress Level vs Physical Activity Level (Fig 3.)', fontsize=12)
plt.xlabel('Physical Activity Level', fontsize=12)
plt.ylabel('Average Stress Level', fontsize=12)
plt.show()
```

Here we explore the relationship between stress and physical activity - another lifestyle choice. More physical activity lowers stress except when one is doing too much physical activity which can lead to overtraining that can cause more stress in the body due to its inability to recover properly. This comparison between stress and exercise is important as stress affects sleep quality. Due to this, one should find ways to decrease their stress. With the data shown, it is clear that with moderate exercise, stress levels decrease. 


Stress Level vs Sleep Duration:

```python
# Scatter plot for stress level vs sleep duration 
sns.scatterplot(data=data, x='Stress Level', y='Sleep Duration', color='dodgerblue', alpha=0.8)
plt.title('Stress Level vs Sleep Duration (Fig 4)', fontsize=12)
plt.xlabel('Stress Level', fontsize=12)
plt.ylabel('Sleep Duration (hours)', fontsize=12)
plt.show()
```

Within this observation, one can see that stress levels have an impact on one's total sleep duration. It is clear that the more stressed an individual is, the less total sleep duration they achieve. 

Stress Levels vs Sleep Quality:

```python
# Bar graph for sleep quality vs stress level
sns.barplot(
    data=data,
    x='Stress Level',
    y='Quality of Sleep',
    errorbar=None,  
    hue='Stress Level',  
    palette='Blues', 
)
plt.title('Average Sleep Quality vs Stress Level (Fig 5.)', fontsize=12)
plt.xlabel('Stress Level', fontsize=12)
plt.ylabel('Average Sleep Quality', fontsize=12)
plt.show()
```

This comparison between stress levels and sleep quality show a pattern in decreased sleep quality with increased stress levels. 

Physical Activity Level vs Sleep Quality:

```python
# Boxplot for sleep quality vs physical acitivty 
sns.boxplot(data=data, x='Physical Activity Level', y='Quality of Sleep')
plt.title('Sleep Quality vs Physical Activity Level (Fig 6.)', fontsize=12)
plt.xlabel('Physical Activity Level (Exercise)', fontsize=12)
plt.ylabel('Quality of Sleep', fontsize=12)
plt.show()
```

This boxplot shows a relationbship between increased sleep quality with moderate excersise with out overtraining. 

Physical Activity Levels vs Sleep Duration:

```python
# Boxplot for sleep duration vs physical acitivity level 
sns.boxplot(data=data, x='Physical Activity Level', y='Sleep Duration')
plt.title('Sleep Duration vs Physical Activity Level (Fig 7.)', fontsize=12)
plt.xlabel('Physical Activity Level (Exercise)', fontsize=12)
plt.ylabel('Sleep Duration (hours)', fontsize=12)
plt.show()
```

This boxplot shows the effect that exercise has on sleep duration. It can be seen that with more exercise (but without overtraining) can increase sleep duration.
